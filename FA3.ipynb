```python
#FAPROGRAM3

name = input("Enter student's name: ")
score = float(input("Enter student's score: "))

if score >= 90 and score <= 100:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print("Student:", name)
print("Grade:", grade)
```
